```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Leer datos
df_dpto = pd.read_csv("..\\data\\processed\\suelo_forestal_depto.csv")
```

```python
df_dpto['Fecha'] = pd.to_datetime(df_dpto['ano'], format='%Y')
df_dpto = df_dpto.sort_values(['depto', 'Fecha'])
```

```python
for depto, subset in df_dpto.groupby('depto'):
    plt.figure(figsize=(8, 4))
    
    plt.plot(subset['Fecha'], subset['total'], marker='o')
    
    plt.title(f"Serie de tiempo - {depto}")
    plt.xlabel("Ao")
    plt.ylabel("Total")
    plt.grid(True)
    
    plt.show()
```

```python
subset = df_dpto[df_dpto['depto'] == depto].sort_values('Fecha')

# Crear ndice de tiempo: 1, 2, ..., 9
subset['t'] = range(1, len(subset) + 1)

# Variable dependiente
Y = subset['total']
# Variable independiente
X = subset[['t']]
```

```python

for depto, subset in df_dpto.groupby('depto'):

    subset = df_dpto[df_dpto['depto'] == depto].sort_values('Fecha')

    # Crear ndice de tiempo:
    subset['t'] = range(1, len(subset) + 1) #  Modelo lineal
    
    subset['t2'] = subset['t'] ** 2 # Modelo cuadratico

    # Variable dependiente
    Y = subset['total']

    # Variable independiente
    X = subset[['t']] # Modelo lineal

    X2 = subset[['t', 't2']] # Modelo cuadratico

    # Crear el modelo
    model_lin = LinearRegression() # Modelo lineal

    model_quad = LinearRegression() # Modelo cuadratico

    # Ajustar el modelo con tus datos
    model_lin.fit(X, Y) # Modelo lineal

    model_quad.fit(X2, Y) # Modelo cuadratico

    y_pred_lin = model_lin.predict(X)
    y_pred_quad = model_quad.predict(X2)

    plt.figure(figsize=(8,5))
    
    plt.plot(subset['t'], Y, color='green', label='Datos reales')

    plt.plot(subset['t'], y_pred_lin, color='yellow', label='Prediccion modelo lineal')
    
    plt.plot(subset['t'], y_pred_quad, color='red', label='Prediccion modelo cuadratico')

    plt.xlabel('Tiempo(t)')
    plt.ylabel('Total')
    plt.title(f'Ajustes de modelos para {depto} : lineal vs cuadratico')
    plt.legend()
    plt.grid(True)
    plt.show()
```
