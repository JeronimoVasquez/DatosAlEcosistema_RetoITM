```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leer datos
df_dpto = pd.read_csv("..\\data\\processed\\suelo_forestal_depto.csv")
```

```python
df_dpto['Fecha'] = pd.to_datetime(df_dpto['ano'], format='%Y')
df_dpto = df_dpto.sort_values(['depto', 'Fecha'])
```

```python
for depto, subset in df_dpto.groupby('depto'):
    plt.figure(figsize=(8, 4))
    
    plt.plot(subset['Fecha'], subset['total'], marker='o')
    
    plt.title(f"Serie de tiempo - {depto}")
    plt.xlabel("Ao")
    plt.ylabel("Total")
    plt.grid(True)
    
    plt.show()
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plot_forecasts(df, log_transform=True):
    for depto, subset in df.groupby('depto'):
        subset = subset.sort_values('Fecha')
        y = subset['total'].copy()
        x = subset['Fecha'].copy()

        # Detectar si log-transform ayuda
        if log_transform:
            y_model = np.log(y)
        else:
            y_model = y

        # Fechas futuras
        future_dates = pd.date_range(x.iloc[-1], periods=4, freq='Y')[1:]

        plt.figure(figsize=(10, 4))
        # Serie real
        plt.plot(x, y, marker='o', label='Serie real')

        # --- Naive ---
        naive_pred = [y_model.iloc[-1]] * 3
        if log_transform:
            naive_pred_plot = np.exp(naive_pred)
        else:
            naive_pred_plot = naive_pred
        plt.plot(future_dates, naive_pred_plot, marker='x', label='Naive')

        # --- SES ---
        try:
            ses = SimpleExpSmoothing(y_model).fit(smoothing_level=0.3, optimized=False)
            ses_pred = ses.forecast(3)
            if log_transform:
                ses_pred_plot = np.exp(ses_pred)
            else:
                ses_pred_plot = ses_pred
            plt.plot(future_dates, ses_pred_plot, marker='x', label='SES')
        except Exception as e:
            print(f"SES fall para {depto}: {e}")

        # Graficar
        plt.title(f"Serie + Predicciones - {depto}")
        plt.xlabel("Ao")
        plt.ylabel("Total")
        plt.grid(True)
        plt.legend()
        plt.show()

# Uso
plot_forecasts(df_dpto)
```
